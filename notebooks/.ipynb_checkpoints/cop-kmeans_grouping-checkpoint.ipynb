### Imports and Params

```python
#imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm #color maps for plotting
import h5py #for reading/writing data
# from copkmeans.cop_kmeans import cop_kmeans #from https://github.com/Behrouz-Babaki/COP-Kmeans/tree/master, installed 5/25/2023
from sklearn.cluster import KMeans
from sklearn import decomposition #for pca
from sklearn.metrics import silhouette_score #for silhouette score
from sklearn.manifold import TSNE #for t-SNE, aids pca
from sklearn.metrics import silhouette_samples, silhouette_score, mean_squared_error #for silhouette score
from time import time
```

```python
#parameters

#file details
volc = 'Hood'
states = 30
norm = 'normalized_' #'normalized_' or ''
h5name = f'/home/smocz/expand_redpy_new_files/h5/new_{norm}{volc.lower()}_templates_{states}_states.h5'

#reduction params
ncom = 3 #number of components | int<4 for t-sne | int<states**2 or 'auto' for pca
idealv = 0.86 #ideal explained variance if ncom='auto'

#t-sne params
init_m = 'pca' #init method 'pca' or 'random', 'pca' is more reproducable
perp = 50 #perplexity, sklearn website recommends 5-50, higher number for larger number of samples
l_rate = 10 #learning rate

#cop-kmeans
k=6
init = 'kmpp'
max_iter = 300 
tol=1e-4
```

```python
#def functions

#getting cluster id from template names as they are saved in the h5 file
if norm == 'normalized_':
    #for normalized template name conventions (CC_BRSP_rpho_12)
    def get_clid(template_byt):
        clid_get = (str(template_byt)[2:-1].split('_')[-1])
        return clid_get
else:
    #for non-normalized template name conventions (brspbhzrpho012)
    if volc == 'St_Helens': #zfill of 4 for helens, 3 for all others
        ii = 4
    else:
        ii = 5
    def get_clid(template_byt):
        clid_get = str(template_byt)[-ii:-1]
        return clid_get

#get ncom from idealv if ncom == 'auto'
def ncom_idealv(ncom_list,idealv,X): #ncom_list is list of options, idealv is ideal variance, x is the data
    sum_list = [] #will be a list of cumulative explained variance ratio for each ncom
    for n_com in ncom_list:
        
        pca = decomposition.PCA(n_components=ncom) #get the pca function
        pca.fit(X) #use pca on data X

        vr = pca.explained_variance_ratio_
        sum_list.append(sum(vr))
        print(f'{ncom} components -> {round(sum(vr),4)} variance ratio')


    diff_list = [abs(i-idealv) for i in sum_list] #find difference from idealv across ncom_list
    #look through ncom_list to find the ncom closest to idealv
    ncom = ncom_list[diff_list.index(min(diff_list))] #get the minimum difference, find its index, pull from ncom_list
    print('auto ncom =',n_com)
    return n_com
```

### Code

```python
#pull in fingerprints
with h5py.File(h5name, "r") as f: #pull in fingerprints
    fingerprints = f["fingerprints"][()]
    template_name = f["template_name"][()]
    waveforms = f["waveforms"][()]
    print(f.keys()) #print what data is in this file
```

```python
#3D to 2D
print('shape of fingerprints',fingerprints.shape)
X = fingerprints.reshape((fingerprints.shape[0], fingerprints.shape[1]**2)) 
print('shape after reshape',X.shape)

#find ncom if needed
if ncom == 'auto':
    ncom = ncom_idealv(ncom_list=[2,5,10,15,20,25,30,35,40,45,50,55,60],idealv=idealv,X=X)
else:
    print(f'ncom = {ncom}')
```

```python
###reduce number of components with pca or t-sne ###

# pca on data X
# pca = decomposition.PCA(n_components=ncom) #get the pca function
# pca.fit(X) #use pca on data X
# X = pca.transform(X) #use pca to transform data X

# print('shape after pca',X.shape)

#t-sne on data X
t0 = time()
X = TSNE(n_components=ncom,init=init_m,perplexity=perp,learning_rate=l_rate).fit_transform(X)
t1 = time()

print('shape after t-SNE',X.shape)
print(f'{t1-t0} seconds for t-SNE')
```

```python
#create must-link constraints across stations

ml = [] #list of tuples of must-link pairs by index

cl_list = np.unique([get_clid(i) for i in template_name])

idx_grp_list = [] #list of lists of group indexes
for cl in cl_list:
    cl_idx_list = [] #will be a list of indexes in a group
    for tname in template_name:
        if get_clid(tname)==cl:
            cl_idx_list.append(list(template_name).index(tname))
    idx_grp_list.append(cl_idx_list)
    for idx in cl_idx_list:
        if idx==min(cl_idx_list):
            continue
        else:
            ml.append((min(cl_idx_list),idx))
    # break

print(ml[-8:])
```

```python
#kmeans
t0 = time()
cop_clusters, cop_centers = cop_kmeans(dataset=X, k=k,ml=ml,initialization=init,max_iter=max_iter,tol=tol)
t1 = time()
print(f'{t1-t0} seconds for COP-Kmeans')

#create dataframe with each fingerprint/waveform info
tdf = pd.DataFrame({"template_name":template_name,"waveform":list(waveforms),"reduced_prints":list(X),
                    "kmeans":list(cop_clusters)})
tdf.head()
```

```python
#get a list of kmeans labels per cluster
cl_kmeans = [] #will be a list of kmeans group labels, same length as cl_list
for cc,cl in enumerate(idx_grp_list): #for each list of idexes (all indexes are one cluster, different stations)
    cl_k_list = [] #will be a list of kmeans group labels for each station of a cluster
    for idx in cl: #for each index of a station in a cluster
        kmeans = tdf.loc[[idx]]["kmeans"].item() #get the kmeans group label
        cl_k_list.append(kmeans) #append it to cl_k_list
    if len(np.unique(cl_k_list))==1: #if there are no conflicting values (list should all be one number repeating)
        cl_kmeans.append(cl_k_list[0]) #append the kmeans group label to cl_kmeans for this cluster
    else: #if there IS a conflict in values, something went wrong with cop kmeans (a constraint must have been violated)
        print(f'constraints violated!! for cluster {cl_list[cc]}')
```

```python
#create dataframe of kmeans labels per cluster
cdf = pd.DataFrame({"cluster_id":cl_list,"kmeans_grp":cl_kmeans}) #"indexes":idx_grp_list,
cdf.head()

#save to csv
# cdf.to_csv(f'tsne_{volc}_k_{k}_cl_grps_{states}_states.csv',index=False)
```

```python
#used for testing while COP-Kmeans is not installed
clusterer = KMeans(n_clusters=k, n_init="auto", random_state=10)
cop_clusters = clusterer.fit_predict(X)
```

```python
#silhouette score
# Create a subplot with 1 row and 2 columns
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.set_size_inches(18, 7)

# The 1st subplot is the silhouette plot
# The silhouette coefficient can range from -1, 1 but in this example all
# lie within [-0.1, 1]
ax1.set_xlim([-0.1, 1])
# The (k+1)*10 is for inserting blank space between silhouette
# plots of individual clusters, to demarcate them clearly.
ax1.set_ylim([0, len(X) + (k + 1) * 10])

# Initialize the clusterer with k value and a random generator
# seed of 10 for reproducibility.
cluster_labels = cop_clusters

# The silhouette_score gives the average value for all the samples.
# This gives a perspective into the density and separation of the formed
# clusters
silhouette_avg = silhouette_score(X, cluster_labels)
print(
    "For n_clusters =",
    k,
    "The average silhouette_score is :",
    silhouette_avg,
)

# Compute the silhouette scores for each sample
sample_silhouette_values = silhouette_samples(X, cluster_labels)

y_lower = 10
for i in range(k):
    # Aggregate the silhouette scores for samples belonging to
    # cluster i, and sort them
    ith_cluster_silhouette_values = sample_silhouette_values[cluster_labels == i]

    ith_cluster_silhouette_values.sort()

    size_cluster_i = ith_cluster_silhouette_values.shape[0]
    y_upper = y_lower + size_cluster_i

    color = cm.nipy_spectral(float(i) / k)
    ax1.fill_betweenx(
        np.arange(y_lower, y_upper),
        0,
        ith_cluster_silhouette_values,
        facecolor=color,
        edgecolor=color,
        alpha=0.7,
    )

    # Label the silhouette plots with their cluster numbers at the middle
    ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

    # Compute the new y_lower for next plot
    y_lower = y_upper + 10  # 10 for the 0 samples

ax1.set_title("The silhouette plot for the various clusters.")
ax1.set_xlabel("The silhouette coefficient values")
ax1.set_ylabel("Cluster label")

# The vertical line for average silhouette score of all the values
ax1.axvline(x=silhouette_avg, color="red", linestyle="--")

ax1.set_yticks([])  # Clear the yaxis labels / ticks
ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])

# 2nd Plot showing the actual clusters formed
colors = cm.nipy_spectral(cluster_labels.astype(float) / k)
ax2.scatter(
    X[:, 0], X[:, 1], marker=".", s=30, lw=0, alpha=0.7, c=colors, edgecolor="k"
)

# Labeling the clusters
centers = clusterer.cluster_centers_
# Draw white circles at cluster centers
ax2.scatter(
    centers[:, 0],
    centers[:, 1],
    marker="o",
    c="white",
    alpha=1,
    s=200,
    edgecolor="k",
)

for i, c in enumerate(centers):
    ax2.scatter(c[0], c[1], marker="$%d$" % i, alpha=1, s=50, edgecolor="k")

ax2.set_title(f"The 2D visualization of the clustered data. (ncom = {ncom})")
ax2.set_xlabel("1st feature/component")
ax2.set_ylabel("2nd feature/component")

plt.suptitle(
    "Silhouette analysis for KMeans clustering on sample data with n_clusters = %d"
    % k,
    fontsize=14,
    fontweight="bold",
)
```
