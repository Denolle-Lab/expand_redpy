```python
#Parameters

#Data obtained from IRIS: http://service.iris.edu/irisws/fedcatalog/docs/1/builder/

#Volcano Lat Lons
volc_lat_lon = {}
volc_lat_lon['Mt_Rainier'] = [46.8528857, -121.7603744]
volc_lat_lon['Mt_Baker'] = [48.7773426,  -121.8132008]
volc_lat_lon['Mt_St_Helens'] = [46.1912, -122.1944]
volc_lat_lon['Mt_Hood']=[45.373221, -121.696509]
volc_lat_lon['Newberry']=[43.7220653, -121.2344654]

#names of volcanoes to get url of and append stations to csv (must appear exactly as it does in volc_lat_lon dictionary)
volcano_list = ['Mt_Rainier','Mt_Baker','Mt_St_Helens','Mt_Hood','Newberry']

#radius in deg (0.450 ~= 50km) (0.27 ~= 30km)
rad = '0.270'

csv_name = '/home/smocz/expand_redpy/csv_catalogs/Volcano_Metadata.csv'
```

```python
#Make csv file, if csv file already exists, this WILL overwrite it
with open(csv_name, 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Volcano_Name','Network','Station','Latitude','Longitude','Elevation','Starttime','Endtime'])
    file.close()
```

```python
for volcano in volcano_list:
    lat = format(round(volc_lat_lon[volcano][0], 3), '.3f')
    lon = format(round(volc_lat_lon[volcano][1], 3), '.3f')
    url = 'http://service.iris.edu/irisws/fedcatalog/1/query?cha=*HZ&starttime=2002-01-01T'+ \
    '00:00:00&endtime=2022-01-01T00:00:00&format=text&includeoverlaps=false&lat='+ \
    str(lat)+'&lon='+str(lon)+'&maxradius='+rad+'&nodata=404'
    print(volcano,url)
```

```python
for volcano in volcano_list:
    lat = format(round(volc_lat_lon[volcano][0], 3), '.3f')
    lon = format(round(volc_lat_lon[volcano][1], 3), '.3f')
    url = 'http://service.iris.edu/irisws/fedcatalog/1/query?cha=*HZ&starttime=2002-01-01T00:00:00&endtime=2022-01-01T00:00:00&format=text&includeoverlaps=false&lat='+str(lat)+'&lon='+str(lon)+'&maxradius='+rad+'&nodata=404'
    #Francesca's Code

    net_sta_dict = {}
    r = requests.get(url)

    # print(r.content.splitlines()[5].split("|")[-1])

    #This for loop extrapolates the necessary data components from the url and adds them to the dictionary
    for line in r.content.splitlines():
        line = line.decode('utf-8')
        if len(line)<1: continue
        if ( "#" not in line ):
            net = line.split("|")[0]
            sta = line.split("|")[1]
            loc = line.split("|")[2]
            cha = line.split("|")[3]
            lat = float(line.split("|")[4])
            lon = float(line.split("|")[5])
            elev = line.split("|")[6] 
            end = line.split("|")[-1]
            start = line.split("|")[-2]
            netsta = net + "." + sta
            if netsta in net_sta_dict: #if the station has already been saved
                a = net_sta_dict[netsta]
                if start < a[5] and end < a[6]: #if new starttime is before old starttime AND the new endtime is before the old starttime
                    net_sta_dict[netsta] = [ net, sta, lat, lon, elev, start, a[6]] #save new starttime but old endtime
                if start < a[5] and end > a[6]: #if new starttime is before old AND the new endtime is after the old
                    net_sta_dict[netsta] = [ net, sta, lat, lon, elev, start, end] #save new starttime and new endtime
                if start > a[5] and end < a[6]: #if new starttime is after old AND the  new endtime is before the old
                    net_sta_dict[netsta] = [ net, sta, lat, lon, elev, a[5], a[6]] #save the old starttime and endtime
                if start > a[5] and end > a[6]: #if new starttime is after old AND the new endtime is after the old
                    net_sta_dict[netsta] = [ net, sta, lat, lon, elev, a[5], end] #save the old starttime but the new endtime
            if netsta not in net_sta_dict:
                net_sta_dict[netsta] = [ net, sta, lat, lon, elev, start, end]
    for i in net_sta_dict:
        if net_sta_dict[i][0] != volcano:
            net_sta_dict[i].insert(0, volcano)
        row = net_sta_dict[i]
        print(row)
        with open(csv_name, 'a', 
                  newline='') as file:
            writer = csv.writer(file)
            writer.writerow(row)
            file.close()
```

```python
#get rid of useless stations

df = pd.read_csv(csv_name)

useless = df[(df['Starttime']<'2009-01-01T00:00:00')&(df['Endtime']<'2010-01-01T00:00:00')].index.tolist()
print(useless)

# for i in useless:
#     print(df.iloc[i]['Starttime'],df.iloc[i]['Endtime'])

df = df.drop(index=useless)
df.to_csv(csv_name, index=False)
```

```python
#also drop stations that weren't active for at least a year

df = pd.read_csv(csv_name)

start_list = df['Starttime'].values.tolist()
end_list = df['Endtime'].values.tolist()
for ii,i in enumerate(start_list):
    dif = UTCDateTime(end_list[ii])-UTCDateTime(start_list[ii])
    if dif < (86400*365):
        useless = df[(df['Starttime']==start_list[ii])&(df['Endtime']==end_list[ii])].index.tolist()
        df = df.drop(index=useless)
        print('dropped',dif/(86400*365),start_list[ii],end_list[ii])
    if dif > (86400*365):
        print('good',dif/(86400*365),start_list[ii],end_list[ii])
#         break
#     break
df.to_csv('~/expand_redpy/csv_catalogs/Volcano_Metadata.csv', index=False)
```

```python
#get list of stations that run continuously 2002-2021, out of curiousity
df = pd.read_csv(csv_name)
# print(df)
rain = []
hele = []
bake = []
hood = []
newb = []
start_list = df['Starttime'].values.tolist()
end_list = df['Endtime'].values.tolist()
for ii,i in enumerate(start_list):
    if start_list[ii] < '2002-01-01T00:00:00' and end_list[ii] > '2022-01-01T00:00:00':
        usefull = df.loc[(df['Starttime'] == start_list[ii]) & (df['Endtime'] == end_list[ii])].index.values.tolist()
#         print(usefull)
        for u in usefull:
            if df.loc[u]['Volcano_Name'] == 'Mt_Rainier':
                rain.append(df.loc[u]['Station'])
            if df.loc[u]['Volcano_Name'] == 'Mt_St_Helens':
                hele.append(df.loc[u]['Station'])
            if df.loc[u]['Volcano_Name'] == 'Mt_Baker':
                bake.append(df.loc[u]['Station'])
            if df.loc[u]['Volcano_Name'] == 'Mt_Hood':
                hood.append(df.loc[u]['Station'])
            if df.loc[u]['Volcano_Name'] == 'Newberry':
                newb.append(df.loc[u]['Station'])
print(len(rain),len(hele),len(bake),len(hood),len(newb))
```
