```python
#see config file for parameter explanations
with open('/home/smocz/expand_redpy/scripts/config.yaml') as file:
    config = yaml.load(file, Loader=yaml.FullLoader)

vv = config['vv']
vv=1
volc_list_names = config['volc_list_names']
volc = volc_list_names[vv]
print(volc)
if volc == 'St_Helens': #for catalog name
    volc_read = 'MountStHelens'
else:
    volc_read = volc

readdir = config['readdir']
homedir = config['homedir']
years = config['years']
```

### NWAC Data

NOTE: no NWAC data for Newberry

```python
weather_csv_list = [f'{homedir}misc/MtBaker-HeatherMeadows_4210_allyears.csv',f'{homedir}misc/MtHoodMeadows-Blue_6540_allyears.csv','newberry',f'{homedir}misc/Paradise_5400_allyears.csv', f'{homedir}misc/MtStHelens-Coldwater_3260_allyears.csv']
weather_csv = weather_csv_list[vv]
print(weather_csv)
```

```python
env_param = 'Temperature  (deg F) ' #the key of the environmental data you want to plot against; see below for keys
# weather_df.keys() #show keys; run after reading in the dataframe
env_p_new = 'AVG Temp (F)' #the key it will be saved as, the average is taken for each day, so keeping "AVG" is recommended
env_csv = 'temp' #how to denote the parameter in the csv name
```

Pull in and organize data. *Can be skipped if already saved to csv*

```python
#open data files

weather_df = pd.read_csv(weather_csv) #weather data from NWAC
weather_df["Date/Time (PST)"] = [pd.to_datetime(i) for i in weather_df["Date/Time (PST)"]] #convert to datetimes
weather_df['Date/Time (UTC)'] = weather_df["Date/Time (PST)"] + pd.Timedelta(hours=8) #convert to UTC

volc_cat = pd.read_csv(f'{readdir}{volc_read}_catalog.csv') #pull REDPy catalog with events

i = 0 #used to determine whether to create or add to dataframe
for f in glob(f'{homedir}events/{volc_list_names[vv]}_*_events.csv'): #pull all backfilled events
    if i==0: #if this is the first pass
        volc_evt = pd.read_csv(f) #create a dataframe of the csv
        i=1 #set i
    else: #if NOT the first pass
        df = pd.read_csv(f) #read csv as dataframe
        volc_evt = pd.concat([volc_evt,df]) #concat with existing dataframe
```

```python
weather_df[:5]
```

Get Daily Average Temperature from weather_df

```python
weather_df.keys() #show keys
```

```python
#find average temperature for each day

weather_df["Date"] = [str(i).split(' ')[0] for i in weather_df["Date/Time (UTC)"]] #add a list of days from the column date/time

avg_temp_list = [] #will be a list of average of a parameter for each row of weather_df

date_trange = trange(len(weather_df["Date"]), desc=f"finding average {env_param}", leave=True) #set range
for d in date_trange: #loop through range
    date = weather_df["Date"][d] #get the date for this index (row of weather_df)
    temp_list = weather_df[weather_df["Date"]==date][env_param].values.tolist() #find all temperatures matching this date
    avg_temp_list.append((sum(temp_list)/len(temp_list))) #average all these temperatures and append
    
weather_df[env_p_new] = avg_temp_list #add average temperatures to weather_df
```

```python
#save each date and the average temp as lists with the same index

day_temps = [] #will be list of average param, same index as unique_days
unique_days = list(np.unique(weather_df['Date'])) #np.unique of date column
for day in unique_days: #for each day
#     print(day)
    day_temps.append(weather_df[weather_df['Date']==day][env_p_new].values.tolist()[0]) #find and append the average temperature
```

combine number of events and temperature into one dataframe

```python
all_cat_dates = []
for cd in volc_cat['datetime']:
    all_cat_dates.append(cd.split('T')[0])
    
cat_dates = list(np.unique(all_cat_dates))

num_cat_dates = []
for cd in cat_dates:
    num = all_cat_dates.count(cd)
    num_cat_dates.append(num)
    
print(num_cat_dates[0], cat_dates[0])
```

```python
#do the same for event catalog
    
all_evt_dates = []
for ce in volc_evt['Earliest_Detection_Time']:
    all_evt_dates.append(ce.split('T')[0])
    
evt_dates = list(np.unique(all_evt_dates))

num_evt_dates = []
for ce in evt_dates:
    num = all_evt_dates.count(ce)
    num_evt_dates.append(num)
    
print(num_evt_dates[0], evt_dates[0])
```

```python
#create a large list of all the dates, from which data will be added based on what is available.

#create list of all dates
all_dates = np.unique(evt_dates+cat_dates+unique_days)


#lists with data for the same index as all_dates
cat_num = [] #number of events per day from the catalog
evt_and_cat_num = [] #number of events per day from the catalog AND backfilled events
env_data = [] #average daily of some environmental parameter

for ad in all_dates:
    if ad in cat_dates: #catalog
        cn = num_cat_dates[cat_dates.index(ad)] #number of events in the catalog
        cat_num.append(cn) #append
    else:
        cn = '0'
        cat_num.append(cn)
        
    if ad in evt_dates: #backfilled + catalog
        evt_and_cat_num.append(num_evt_dates[evt_dates.index(ad)]+int(cn)) #add sum
    else:
        evt_and_cat_num.append(cn) #add catalog
        
    if ad in unique_days: #temperature
        env_data.append(day_temps[unique_days.index(ad)])
    else:
        env_data.append('N/A')

plot_df = pd.DataFrame({'Date (UTC)':all_dates,'REDPy_Events':cat_num,'REDPy_and_Backfilled':evt_and_cat_num,env_p_new:env_data})
```

### CDO Data

Stations:
* USC00358246 == SUNRIVER, OR US == ~15 mi from Newberry
* USC00456898 == RAINIER PARADISE RANGER STATION, WA US == Rainier
* USC00353402 == GOVERNMENT CAMP, OR US == Mt Hood

Acronym Guide:

* PRCP = Precipitation
* SNOW = Snowfall
* SNWD = Snow depth
* TMAX = Maximum temperature
* TMIN = Minimum temperature

Units: Imperial (Inches, Fahrenheit, etc.)

Time should be UTC (<a href="https://www.ncei.noaa.gov/products/land-based-station/global-historical-climatology-network-daily">"The Global Summary of the Day contains 24-hour summaries encoded in the special “climatological code” group transmitted with synoptic (SYNOP) reports on the Global Telecommunications System (GTS). NCEI archives these reports in the Integrated Surface Dataset (ISD; DSI-3505) and the 24-hour summary period generally ends at midnight UTC (i.e., 2400 UTC)."</a>

see <a href="https://docs.opendata.aws/noaa-ghcn-pds/readme.html">this page</a> for more information on acronyms for GHCND (Global Historical Climatology Network Daily)

```python
cdo_df = pd.read_csv(f'{homedir}/misc/CDO_data.csv')
cdo_df = cdo_df.drop(axis=1,labels=['DAPR','DASF','MDPR','MDSF','WT01','WT03','WT04','WT05','WT06','WT11']) #drop empty columns

sunriver_df = cdo_df[cdo_df['STATION']=='USC00358246']
paradise_df = cdo_df[cdo_df['STATION']=='USC00456898']
govcamp_df = cdo_df[cdo_df['STATION']=='USC00353402']

cdo_df = govcamp_df #select which dataframe (station data) to use, depending on volcano
```

```python
summer_months = list(np.arange(4,10,1)) #apr-sep (winter is then oct-mar)

fig, axs = plt.subplots(nrows=2, ncols=1,figsize = (10,5))

i = 0
for ax in axs:
#     print(i)
    if i==0: #if first axis
        
        x = [pd.to_datetime(d) for d in list(plot_df['Date (UTC)'].values)] #x is dates
        y = list(plot_df['REDPy_and_Backfilled'].values) #y is total number of events

        ax.grid(which='major')
        ax.grid(which='minor',color='#EEEEEE')

        for xx,yy in zip(x,y): #plot total events (taller lines)
            if xx.month in summer_months: #if its in summer
                ax.vlines(xx,ymin=0,ymax=yy, color='#7B0004')
            else: #if its in winter
                ax.vlines(xx,ymin=0,ymax=yy, color='#416D9F')
                
        y = list(plot_df['REDPy_Events'].values) #y is now JUST REDPy events
                
        for xx,yy in zip(x,y):
            if xx.month in summer_months: #if its in summer
                ax.vlines(xx,ymin=0,ymax=yy, color='red')
            else: #if its in winter
                ax.vlines(xx,ymin=0,ymax=yy, color='#BFD7EA')
        
        ax.set_xlim(left = min(x),right=max(x))
#         ax.set_ylim(bottom=0,top=5)
        ax.set_ylabel('Number of Events Per Day')

        
    if i==1: #if second axis
        

        ax.plot(x,[float(pn) for pn in list(plot_df[env_p_new].values)],color='black')
        
        ax.set_xlabel('Date (UTC)')
        ax.set_ylabel(f'Daily {env_param}')

        ax.set_xlim(left = min(x),right=max(x))
        
        ax.grid(which='major')
        ax.grid(which='minor',color='#EEEEEE')
        
    i=i+1

fig.suptitle(f'{volc} Events per Day against {env_param}')
        
```
