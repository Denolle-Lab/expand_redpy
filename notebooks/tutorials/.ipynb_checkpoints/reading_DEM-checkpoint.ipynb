### 1. Download DEM data

#### There are 2 ways depending on the resolution you want (10m for closeups, or 61+m for larger maps)

##### 10m data:

For Washington State 10m data:
<a href="http://gis.ess.washington.edu/data/raster/tenmeter/byquad/index.html"> gis.ess.washington.edu raster data </a>
1. select area (ex: Yakima)
2. you will be taken into the files for that area, click on a square on the map to download the dem file. Or, scroll down and click the "DNR number" for the area you want to download it.
3. unzip the dem file
4. THESE DEM FILES ARE IN UTM (NOT LAT LON) AND NEED TO BE CONVERTED TO .TIF FILE FOR THIS METHOD

    a. method for converting to .tif file will be added eventually

##### 61+m data:

Any location:
<a href="https://gmrt.org/GMRTMapTool/"> gmrt.org MapTool </a>
1. Navigate to desired location
2. Click and drag to select an area to get DEM from. Release click to stop.

OR

1. Set maximum and minimum lat and lon in the box at the center right
2. On keyboard, hit "Enter" or "Return" to save that box, check on the map that your boundaries look correct

3. Select "Create Grid File" at the center right
4. Select "GeoTIFF" file format
5. "Unmasked" is fine
6. Select desired resolution, the resolutions available will vary by region
7. Click "Download Grid"
8. Save file and put it in the directory you want.


### 2. Imports

```python
#import things
import numpy as np #numpy for math
import rasterio as rio #rasterio for reading in DEM
from rasterio.plot import show #for plotting with rio
import matplotlib.pyplot as plt #for plotting (base of plotting)
from pyproj import Proj,transform,Geod #for converting to UTM
import pandas as pd #pandas for reading in csvs when plotting locations
```

### 3. Open and Show DEM (from .tif file)

This example is from gmrt.org, with 61m resolution

```python
dem = rio.open("/home/smocz/expand_redpy_new_files/DEM_data/St_Helens_lessbig.tif") #open tif file (contains DEM) with rio
dem_array = dem.read(1).astype('float64') #set array for DEM
crs = dem.crs #get crs (coordinate reference systems) for going between lat lon and UTM
dem_data_dict={'data':dem_array, 'crs':crs, 'left':dem.bounds[0], 'right':dem.bounds[2], 'bottom':dem.bounds[1],
               'top':dem.bounds[3]} #extract boundaries
```

```python
#basic plot can now be made
data = dem_data_dict['data']
left,right,bottom,top = dem_data_dict['left'],dem_data_dict['right'],dem_data_dict['bottom'],dem_data_dict['top']

fig, ax = plt.subplots(1, figsize=(12,12)) #make plot

#plot DEM data, extent sets axis values correctly, cmap is colors
ax.imshow(data,extent=[left, right, bottom, top],cmap='Greys_r') 

plt.show() #show plot
```

### 4 Plot Points on the DEM

Includes transforming points from lat lon to UTM

```python
#can make a dictionary of lat_lons to plot various points easy, in this case volcano centers
volc_lat_lon = {'Rainier': [46.8528857, -121.7603744], 'Baker': [48.7773426, -121.8132008], 
                'St_Helens': [46.1912, -122.1944], 'Hood': [45.373221, -121.696509], 
                'Newberry': [43.7220653, -121.2344654]} 
```

```python
#pull in data, mine is in csv and the above dictionary

#find center point
center_lat = volc_lat_lon['St_Helens'][0]
center_lon = volc_lat_lon['St_Helens'][1]

#pull in template locations
loc_df = pd.read_csv(f'/home/smocz/expand_redpy_new_files/locations/St_Helens_Template_Locations.csv') #get locations
loc_lats = loc_df['Latitude'].values.tolist() #get a list of latitudes for template locations
loc_lons = loc_df['Longitude'].values.tolist() #get a list of longitudes for template locations

#pull station locations
volc_md = pd.read_csv(f'/home/smocz/expand_redpy/csv_catalogs/Volcano_Metadata.csv') #read metadata
stas_lats = volc_md[volc_md['Volcano_Name']=='St_Helens']['Latitude'].values.tolist() #get latitude for stations at this volcano
stas_lons = volc_md[volc_md['Volcano_Name']=='St_Helens']['Longitude'].values.tolist() #get latitude for stations at this volcano
```

```python
#transform locations and stations to UTM for 10M DATA, DOES NOT NEED TO BE DONE FOR GMRT TIFF
#can take a few minutes depending on amount

#get data for conversion and units
crs = dem_data_dict['crs'] #get crs
data = dem_data_dict['data'] #get data
p2 = Proj(crs,preserve_units=False) #set up crs
p1 = Proj(proj='latlong',preserve_units=False) #set up for latlong

loc_eastings = [] #eastings for locations, from longitude
loc_northings = [] #northings for locations, from latitude
stas_eastings = [] #eastings for stations, from longitude
stas_northings = [] #northings for stations, from latitude

for ll,l in enumerate(loc_lats): #for each location, loc_lons(ll) is lon, l is lat
#     print(f' loc lon_lat: {loc_lons[ll]}, {l}')
    loc_easting, loc_northing = transform(p1,p2,loc_lons[ll],l) #transform center of volcano to UTM
#     print(f' loc UTM: {loc_easting}, {loc_northing}')
    loc_eastings.append(loc_easting) #append to list
    loc_northings.append(loc_northing) #append to list
    
for ll,l in enumerate(stas_lats): #for each station, stas_lons(ll) is lon, l is lat
    sta_easting, sta_northing = transform(p1,p2,stas_lons[ll],l) #transform center of volcano to UTM
    stas_eastings.append(sta_easting) #append to list
    stas_northings.append(sta_northing) #append to list
```

```python
print(f'left: {left} | right: {right} | bottom: {bottom} | top: {top}')
print(f'center: {center_lon}, {center_lat}')

fig,ax = plt.subplots(1,1,figsize=(12,12)) #make figure
ax.imshow(data,extent=[left, right, bottom, top],cmap='gist_earth') #plot DEM
ax.set_title('Event and Station Locations on Mt St Helens') #title
ax.scatter(loc_lons, loc_lats,marker='.',color='red',label='Template Locations') #plotting some points
ax.scatter(stas_lons, stas_lats,marker='o',color='white',label='Stations') #plotting some points
ax.scatter(center_lon, center_lat,marker='*',color='black',s=100,label='Volcano Center') #plotting volcano center
ax.legend() #show legend

#crop/zoom-in on map
ax.set_xlim(left+.2,right-0.2) #set manually to what looks best
ax.set_ylim(bottom+0.05,top-.15) #set manually to what looks best
```
